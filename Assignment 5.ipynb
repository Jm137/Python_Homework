다음은 내장 자료형 list를 서브클래싱하여 만든 MySet 클래스 정의 내용이다. 다음 클래스 정의에서 __init__(), __str()__(), elimicate_duplicate()의 세 개의 메소드 코드 내용을 자신이 다른 사람에게 가르친다고 생각하며 설명해보시오.

MySet은 집합(Set) 자료형을 정의하려는 의도하에 만들어진 클래스이다.

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print s
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print t
```

먼저, MySet 클래스는 list 클래스를 상속 받아 만들어진 자식 클래스입니다.

1. init() : 생성자에서 인자로 받은 리스르 l에 있는 리스트 구성원들을 for문을 통해 하나씩 받아와 넣어 줍니다.(append) 이후, MySet 클래스의 eliminate_duplicate 메소드를 호출합니다.

2. eliminate_duplication() : staticmethod(정적메소드)는 인스턴스 객체와 무관하게 클래스 이름 공간에 존재하는 메소드로서 클래스 이름을 이용하여 직접 호출할 수 있는 메소드를 말합니다. 따라서 init() 즉, 생성자에서 객체 생성 없이 호출 될 수 있었습니다. 동작으로는 먼저 인자로부터 받아 온 리스트(l)로 부터 새로운 리스트 s에 for문을 이용해 넣어주는데, 단 if e not in s 인 경우(e가 리스트 s에 없는 경우)만 넣어줍니다. 그리고 다시 for문을 이용하여 인자로 받아왔던 리스트(l)을 초기화 해준 후 다시 넣어 줍니다.

3. str() : str 함수는 print문을 이용해 출력해 줄 때, 호출이 되는 함수로 for문을 이용해 result 안에 " 원소값 + ',' "  를 계속 해서 넣어 주고 for문이 끝난 후에는 마지막 ','를 제거하고 "}"를 넣어 주고, result 값을 반환해 준다.

[6번 문제]


5번 문제에서 정의한 MySet 클래스에 대해 다음 예제를 수행하면 오류없이 올바르게 동작하는 것을 확인할 수 있다. 다음 예제 내에 있는 len(), bool() 내장함수와 in 키워드 사용 예제가 별다른 메소드 정의를 하지 않았는 데도 올바르게 수행되는 이유를 설명하시오.

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
            
    # 두 집합의 합집합을 반환        
    def __or__(self, l):
        for e in l:
            if e not in self:
                self.append(e)
        return MySet(self)
    
    # 두 집합의 교집합을 반환
    def __and__(self, l):
        s = []
        for e in self:
            if e in l:
                s.append(e)
        return MySet(s)
    
    # 두 집합의 차집합을 반환
    def __sub__(self, l):
        for e in self:
            if e in l:
                self.remove(e)
        return MySet(self)
        
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print s
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print t
print

print '두 집합의 합집합을 반환' 
u = s | t
print u
print

print '두 집합의 교집합을 반환'
s = MySet([1, 2, 2, 3])
t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
k = s & t
print k
print

print '두 집합의 차집합을 반환'
s = MySet([1, 2, 3])
t = MySet([3, 4, 5])
u = s - t
print u
print

###########################################################################
print '6번 문제'
print

s = MySet([1, 2, 3, 4, 5, 6])
print len(s)
#6

print bool(s)
#True

print 2 in s
#True
```

MySet 클래스가 list 클래스를 상속 받은 자식 클래스이기 때문이다. 따라서 리스트의 길이를 반환하는 len() 함수, 리스트가 비어있는지를 반환하는 bool() 함수, 리스트에 해당 원소가 들어 있는지를 반환하는 in 등을 호출하여 사용할 수 있다.

[11번 문제]

n! 이라는 표기법은 n × (n − 1) × ... × 3 × 2 × 1을 뜻합니다.

예를 들자면 10! = 10 × 9 × ... × 3 × 2 × 1 = 3628800 이 되는데,
여기서 10!의 각 자리수를 더해 보면 3 + 6 + 2 + 8 + 8 + 0 + 0 = 27 입니다.

100! 의 자리수를 모두 더하면 얼마입니까?

```python
def factorial(n):
    new = 1
    final = 0
    for x in range(1, n + 1):
        new = x * new
    while new:           
        digit = new % 10            
        new //= 10
        final += digit
    return (final)

print (factorial(100))
```

[12번 문제]

n의 약수들 중에서 자신을 제외한 것의 합을 d(n)으로 정의했을 때,
서로 다른 두 정수 a, b에 대하여 d(a) = b 이고 d(b) = a 이면 
a, b는 친화쌍이라 하고 a와 b를 각각 친화수(우애수)라고 합니다.

예를 들어 220의 약수는 자신을 제외하면 1, 2, 4, 5, 10, 11, 20, 22, 44, 55, 110 이므로 그 합은 d(220) = 284 입니다.
또 284의 약수는 자신을 제외하면 1, 2, 4, 71, 142 이므로 d(284) = 220 입니다.
따라서 220과 284는 친화쌍이 됩니다.

10000 이하의 친화수들을 모두 찾아서 그 합을 구하세요.

```python
sum1 = sum2 = total = d1 = d2 = 0
nList = range(1,10001)
for n in nList:
    sum1 = sum2 = 0
    for d1 in range(1,n):
        if n % d1 == 0:
            sum1 += d1
    if sum1 < 10001 and sum1 != n:
        for d2 in range(1,sum1):
            if sum1 % d2 == 0:
                sum2 += d2
        if d1+1 == sum2 and d2+1 == sum1:
            total += sum1 + sum2
print total/2
```

```python
# -*- coding: utf-8 -*-
import urllib2

def tag_delete(string):
    for i in range(0, string.count('<!--')):
        if string.find("<!--") != -1:
            string = string.replace(string[string.find("<!--"):string.find("-->")+3:]," ")
    for i in range(0,string.count("<")):
        if string.find("<") != -1:
            string = string.replace(string[string.find("<"):string.find(">")+1:]," ")
    
    return string

#자바스크립트 제거
def javascript_delete(string):
    for i in range(0, string.count('<script')):
        if string.find("<script") != -1:
            string = string.replace(string[string.find("<script"):string.find("</script>")+9:]," ")
    return string

#css 제거
def stylesheet_delete(string):
    for i in range(0, string.count('<style')):
        if string.find("<style") != -1:
            string = string.replace(string[string.find("<style"):string.find("</style>")+8:]," ")
    return string

source = stylesheet_delete(source)
source = javascript_delete(source)
source = tag_delete(source)

import string
#구두문자 제거 함수
def punc_delete(source):
    punctuation = string.punctuation
    for i in range (len(punctuation)):
        source= source.replace(punctuation[i],"")
        
    return source


#불용어 제거 함수
def stopword_delete(words):
    stopword = [ 'A','a', 'and', 'are', 'as', 'at', 'be', 'but','But', 'by',
                'for', 'if','If', 'in', 'into', 'is', 'it','It', 'no','No',
                'not','Not', 'of', 'on', 'or', 's', 'such', 't', 'that','That',
                'the','The', 'their','Their', 'then','Then', 'there','There',
                'these','These', 'they','They', 'this','This', 'to', 'was', 'will',
                'with']
    for i in range (len(stopword)):
        for j in range(len(words)):
            if words[j] == stopword[i]:
                words[j] = ""
                
source = punc_delete(source)
words = source.split()
                
#사전형태로 출현빈도 저장 함수
def word_count(words):
    words_dict = {}
    for i in words:
        if words_dict.has_key(i) is False:
            words_dict[i]=1
        else:
            words_dict[i]+=1
    return words_dict

words = word_count(words)


class WebWordsFrequency(list):
    def __init__(self, *args):
        for mem in args:
            self.append(mem)
            
    def addUrl(self, other):
        self.append(other)
        return WebWordsFrequency(self)
    
    def removeUrl(self, other):
        self.remove(other)
        return WebWordsFrequency(self)
        
    def listUrls(self):
        print self
    
    def getWordsFrequency(self):
        for mem in self:
            source = urllib2.urlopen(mem).read()
            source = stylesheet_delete(source)
            source = javascript_delete(source)
            source = tag_delete(source)
            source = punc_delete(source)
            words = source.split()
            stopword_delete(words)
            words = word_count(words)
            
        for i in words:
            print str(i) + " : " + str(words[i])
    
    def getMaxFrequencyWords(self):
        for mem in self:
            source = urllib2.urlopen(mem).read()
            source = stylesheet_delete(source)
            source = javascript_delete(source)
            source = tag_delete(source)
            source = punc_delete(source)
            words = source.split()
            stopword_delete(words)
            words = word_count(words)
       
        combine = {} 
        combine = words
        g = combine.items()
        
        g.sort(key=lambda x:x[1],reverse=True)
        print "제일 많이 나온 것"
        print g[1]
        
class OrderedWebWordsFrequency(WebWordsFrequency):
    def getWordsFrequency(self):
        self.reverse = 'False'
    
        for mem in self:
            source = urllib2.urlopen(mem).read()
            source = stylesheet_delete(source)
            source = javascript_delete(source)
            source = tag_delete(source)
            source = punc_delete(source)
            words = source.split()
            stopword_delete(words)
            words = word_count(words)
            
        combine = {} 
        combine = words
        g = combine.items()
        
        g.sort(key=lambda x:x[1],reverse = True)
     
        for i in range(len(combine)):
            print g[i]
    
    
w1 = WebWordsFrequency('http://www.cnn.com', 'http://www.times.com', 'https://www.amazon.com') 
w2 = WebWordsFrequency('http://www.cnn.com', 'http://www.times.com')
w3 = WebWordsFrequency() 

w1.addUrl('https://github.com')
w3.addUrl('http://stackoverflow.com')

w1.removeUrl('http://www.cnn.com')
w2.removeUrl('http://www.cnn.com')
w3.removeUrl('http://stackoverflow.com')

w1.listUrls()
w2.listUrls()
w3.listUrls()

w2.getWordsFrequency()
w2.getMaxFrequencyWords() 


w4 = OrderedWebWordsFrequency('http://www.times.com')
w4.getWordsFrequency()
w4.getWordsFrequency(reverse=true) 
    
```
